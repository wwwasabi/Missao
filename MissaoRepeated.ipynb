```python
from sklearn.model_selection import RepeatedKFold
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

dados = pd.read_csv('glass.data', names= ["Id number", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type of glass"])
dados.drop('Id number', axis=1, inplace=True)
y = dados['Type of glass']
x = dados.drop('Type of glass', axis = 1)

#Normalizando os dados de entrada
normalizadorMinMaxScaler = MinMaxScaler(feature_range=(0, 1))
x = normalizadorMinMaxScaler.fit_transform(x)
```

```python
rkf = RepeatedKFold(n_splits=5, n_repeats=10, random_state=2652124)
```

```python
for train_index, test_index in rkf.split(x):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```
