```python
from sklearn import svm
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import pandas as pd


dados = pd.read_csv('glass.data', names= ["Id number", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type of glass"])
dados.drop('Id number', axis=1, inplace=True)
y = dados['Type of glass']
x = dados.drop('Type of glass', axis = 1)
```

```python
#Normalizando os dados de entrada
normalizadorMinMaxScaler = MinMaxScaler(feature_range=(0, 1))
x_norm = normalizadorMinMaxScaler.fit_transform(x)
```

```python
#Separando os dados em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x_norm, y, test_size=0.3, random_state=16)
```

```python
#Treinando o modelo SVM
linear = svm.SVC(kernel='linear', C=9, decision_function_shape='ovo').fit(x_train, y_train)

#Analisando as métricas
linear_pred = linear.predict(x_test)
accuracy_lin = linear.score(x_test, y_test)
print("Predição:", linear_pred)
print("Acurácia:", accuracy_lin)
```

```python
#Aplicando a Matriz Confusão
cm_lin = confusion_matrix(y_test, linear_pred)
print(cm_lin)
```
