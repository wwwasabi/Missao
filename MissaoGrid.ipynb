```python
from sklearn import svm
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
import pandas as pd
import numpy as np


dados = pd.read_csv('glass.data', names= ["Id number", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type of glass"])
dados.drop('Id number', axis=1, inplace=True)
y = dados['Type of glass']
x = dados.drop('Type of glass', axis = 1)

#Normalizando os dados de entrada
normalizadorMinMaxScaler = MinMaxScaler(feature_range=(0, 1))
x = normalizadorMinMaxScaler.fit_transform(x)
```

```python
#Criando atributos Grid para SVM
kernel1 = ['linear', 'rbf', 'poly', 'sigmoid']
C1 = np.array([1, 2, 3 , 4, 5, 10])
decision_function_shape1 = ['ovo', 'ovr']

#Dicionario com todos os valores
valores = {'kernel':kernel1, 'C':C1, 'decision_function_shape':decision_function_shape1}


#Criando o modelo SVM e aplicando a Grid com CV
modelo = svm.SVC()
grid_modelo = GridSearchCV(estimator=modelo, param_grid=valores, cv=5)
grid_modelo.fit(x, y)

#Melhores valores para SVM nessa base de dados
print('Melhor Score:', grid_modelo.best_score_)
print('Kernel: ', grid_modelo.best_estimator_.kernel)
print('Degree:', grid_modelo.best_estimator_.degree)
print('Gamma:', grid_modelo.best_estimator_.gamma)
print('Coef0: ', grid_modelo.best_estimator_.coef0)
print('C: ', grid_modelo.best_estimator_.C)
print('Shrinking: ', grid_modelo.best_estimator_.shrinking)
print('probability: ', grid_modelo.best_estimator_.probability)
print('Verbose: ', grid_modelo.best_estimator_.verbose)
print('DFS: ', grid_modelo.best_estimator_.decision_function_shape)
print('Break Ties: ', grid_modelo.best_estimator_.break_ties)
```
